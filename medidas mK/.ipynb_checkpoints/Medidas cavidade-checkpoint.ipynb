# Cavidade $\lambda/4$

```python
from instruments.E5063A_driver import *
import numpy as np
import time
import matplotlib.pyplot as plt
from scipy.signal import find_peaks, peak_prominences

try:
    vna = E5063A_driver("TCPIP0::169.254.101.105::inst0::INSTR")
except:
    print("Comm Er")
#Controle do atenuador
from instruments.Agilent11713C_driver import * #attenuator
att = Agilent11713C_driver('TCPIP0::169.254.101.101::inst0::INSTR')
#funções que da pra usar: att.get_attenuation() pra obter uma atenuação
#                         att.set_attenuation(valor) pra definir uma atenuação (coloca positivo mesmo)

def remove_odd_terms(lst):
    return [x for x in lst if x % 2 != 0]

import measurements.powersweep as powersweep
```

```python
#REALIZA MEDIDAS
def measure(start, stop, sleep, if_bandwidth, power=-5): # Inserir os valores em GHz, tempo em segundos, IF bandwidth em Hz, power entre 0 e -20 dBm
    start = start*1e9
    stop = stop*1e9
    f_c = (stop+start)/2
    span_f = stop-start
    n_pontos = 10000
    
    vna.power = 0
    vna.clear_averaging()
    
    vna.average_points = 128
    vna.averaging = 1
    #vna.set_amplitude(power)
    
    vna.center_frequency = f_c
    vna.span_frequency = span_f
    vna.if_bandwidth_frequency = if_bandwidth
    vna.sweep_points = n_pontos
    vna.autoscale(1)
    
    vna.power = 1
    
    time.sleep(sleep)
    
    freq = np.arange( (f_c-(span_f/2))/1e9, (f_c+(span_f/2))/1e9, (span_f/1e9)/n_pontos)
    cont = 0
    x = []
    for i in freq:
        x.append(freq[cont])
        cont = cont + 1
    if len(x) == n_pontos+1:
        x.pop()
    data = np.array(vna.data_array().split(',')).astype(float)
    data = remove_odd_terms(data)
    plt.plot(x,data)
    
    # Adicionar título e rótulos dos eixos
    plt.title('S21')
    plt.ylabel('Transmissão [dB]')
    plt.xlabel('Freq [GHz]')
    plt.show()
    #vna.power = 0

    return [x, data]

# SALVA DOS DADOS EM .TXT
def save_data(nome_arquivo, eixo_x, eixo_y): # inserir o nome do arquivo (exemplo.txt)
    # Fazer uma matriz com os valores dos eixos. matriz = [eixo x, eixo y] -> fica no formato de linhas
    x = eixo_x
    data = eixo_y
    matriz = [x, data]
    # faz a transposta da matriz pra obter no formato de colunas
    M_t = np.array(matriz).T
    
    # recebe um nome para salvar os dados
    nome_arquivo = nome_arquivo
    
    # Salvando os dados em documento .txt
    with open(nome_arquivo,'w') as arquivo:
        for valor in M_t:
            arquivo.write(str(valor)+'\n')

# FAZ O PLOT DO GRAFICO
# Pra plotar o gráfico desse jeito, os dados do arquivo .txt devem estar no formato [numero    numero]
def plot(nome_arquivo, retornar_valores = False): # Importar o arquivo .txt criado e plotar o gráfico
    dataset = open(nome_arquivo,'r')
    u = [] #eixo x
    v = [] #eixo y
    for line in dataset:
        line = line.replace('[',' ') #troca o simbolo [ por um espaço
        line = line.replace(']',' ') #troca o simbolo ] por um espaço
        line = line.strip() #elimina todos os espaços no início e no final da lista
        U, V = line.split() #sepada os dois valores em duas variaveis diferentes
        u.append(U)
        v.append(V)
    
    dataset.close()
    
    # Até aqui são strings, vou transformar em float
    lista_u = []
    lista_v = []
    j = 0
    for i in u: #transformando o eixo x em uma lista de números
        k = float(u[j])
        lista_u.append(k)
        j=j+1
    j = 0
    for i in v: #transformando eixo y em uma lista de números
        k = float(v[j])
        lista_v.append(k)
        j=j+1
    
    plt.plot(lista_u, lista_v)

    nome_grafico = nome_arquivo.replace('.txt',' ')
    nome_grafico = nome_grafico.replace('.',',')
    nome_grafico = nome_grafico.replace('_',' ')
    
    # Adicionar título e rótulos dos eixos
    plt.title(nome_grafico)
    plt.ylabel('Transmissão [dB]')
    plt.xlabel('Freq [GHz]')
    #plt.show()

    if retornar_valores == True:
        return [lista_u,lista_v]

# SALVAR A FIGURA DO GRAFICO    
def salva_imagem(nome_arquivo): # Importar o arquivo .txt criado e salvar o gráfico
    plt.close()
    dataset = open(nome_arquivo,'r')
    u = [] #eixo x
    v = [] #eixo y
    for line in dataset:
        line = line.replace('[',' ') #troca o simbolo [ por um espaço
        line = line.replace(']',' ') #troca o simbolo ] por um espaço
        line = line.strip() #elimina todos os espaços no início e no final da lista
        U, V = line.split() #separa os dois valores em duas variaveis diferentes
        u.append(U)
        v.append(V)
    
    dataset.close()
    
    # Até aqui são strings, vou transformar em float
    lista_u = []
    lista_v = []
    j = 0
    for i in u: #transformando o eixo x em uma lista de números
        k = float(u[j])
        lista_u.append(k)
        j=j+1
    j = 0
    for i in v: #transformando eixo y em uma lista de números
        k = float(v[j])
        lista_v.append(k)
        j=j+1
    plt.plot(lista_u, lista_v)
    nome_grafico = nome_arquivo.replace('.txt',' ')
    nome_grafico = nome_grafico.replace('.',',')
    nome_grafico = nome_grafico.replace('_',' ')
    
    # Adicionar título e rótulos dos eixos
    plt.title(nome_grafico)
    plt.ylabel('Transmissão [dB]')
    plt.xlabel('Freq [GHz]')
    #plt.show()
        
    plt.savefig(nome_grafico) # Salva a imagem
    plt.close()

# ENCONTRA OS PICOS NO GRAFICO
def encontra_picos(nome_arquivo, prominencia, retornar_valores = False):
    eixox, eixoy = plot(nome_arquivo, True)

    picos, _ = find_peaks(eixoy) # encontra os picos

    prominencias = peak_prominences(eixoy, picos)[0] # calcula a prominencia dos picos

    limite_prominencia = prominencia # define um limite de prominencia para filtrar picos menores

    picos_filtrados = [pico for pico, prom in zip(picos, prominencias) if prom > limite_prominencia] # encontra picos com uma prominencia acima do limite

    eixoy = np.array(eixoy) # transforma lista em np.array
    eixox = np.array(eixox)

    plt.plot(eixox, eixoy, color='blue')
    #plt.scatter(eixox[picos], eixoy[picos], color='black')
    plt.scatter(eixox[picos_filtrados], eixoy[picos_filtrados], color='red')
    plt.show()

    if retornar_valores == True:
        return [eixox[picos_filtrados],eixoy[picos_filtrados]]
        
```
