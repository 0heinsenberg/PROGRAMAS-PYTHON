```python
def ler_dados(nome_arquivo):
    dados_x = []
    dados_y = []
    try:
        with open(nome_arquivo, 'r') as arquivo:
            for linha in arquivo:
                valores = linha.strip('[]\n').split()
                x = float(valores[0])
                y = float(valores[1])
                dados_x.append(x)
                dados_y.append(y)
    except FileNotFoundError:
        print("O arquivo n√£o foi encontrado.")
    except Exception as e:
        print("Ocorreu um erro:", e)
    return dados_x, dados_y
```

```python
def ler_csv(nome_arquivo):
    dados = []
    try:
        with open(nome_arquivo, 'r', newline='') as arquivo_csv:
            leitor_csv = csv.reader(arquivo_csv)
            for linha in leitor_csv:
                # Dividir a linha pelo espa√ßo e ponto e v√≠rgula
                informacoes = linha[0].split(';')
                segunda_informacao = informacoes[0]
                terceira_informacao = informacoes[1]
                # Ignorar a primeira e √∫ltima informa√ß√£o
                dados.append((segunda_informacao, terceira_informacao))
    except FileNotFoundError:
        print("O arquivo n√£o foi encontrado.")
    except Exception as e:
        print("Ocorreu um erro:", e)
    return dados

def plotar_grafico_dB(data):
    try:
        # Separar os valores de x e y
        x = [float(item[0])/(10**9) for item in data]
        y = [float(item[1]) for item in data]

        # Plotar os valores
        plt.plot(x, y)
        plt.xlabel('Frequ√™ncia [GHz]')
        plt.ylabel('Transmiss√£o[dB]')
        plt.title(r'Gr√°fico FUNDO ELSA em dB')
        plt.show()
    except Exception as e:
        print("Ocorreu um erro ao plotar o gr√°fico:", e)

        
        for k in range (0,len(dados_y)):
            calc_y = 10 ** (10/dados_y[k])
            y_pot.append(calc_y)
    
def plotar_grafico_W(data):
    try:
        # Separar os valores de x e y
        x = [float(item[0])/(10**9) for item in data]
        y = [(10**(10/(float(item[1])))) for item in data]

        # Plotar os valores
        plt.plot(x, y)
        plt.xlabel('Frequ√™ncia [GHz]')
        plt.ylabel('Transmiss√£o [W]')
        plt.title(r'Gr√°fico FUNDO ELSA em W')
        plt.show()
    except Exception as e:
        print("Ocorreu um erro ao plotar o gr√°fico:", e)
```

```python
import matplotlib.pyplot as plt

def plot_com_eixo_omitido(x_data, y_data, x_omitido_index, x_omitido_value):
    plt.plot(x_data, y_data, marker='o', linestyle='-')
    plt.xticks(x_data + [x_omitido_value])
    plt.xlabel('Eixo X')
    plt.ylabel('Eixo Y')
    plt.title('Plot com eixo X omitido')
    
    # Adicionando marcador para o valor omitido
    plt.annotate(f'Omitido: {x_omitido_value}', xy=(x_omitido_index, y_data[x_omitido_index]), xytext=(x_omitido_index + 0.5, max(y_data) * 0.8),
                 arrowprops=dict(facecolor='black', arrowstyle='->'),
                 )
    
    plt.show()

# Exemplo de dados
x_data = [1, 2, 3, 5, 6, 7, 8]
y_data = [4, 7, 2, 5, 8, 9, 3]
x_omitido_index = 3  # √çndice do valor a ser omitido
x_omitido_value = 4   # Valor a ser omitido

plot_com_eixo_omitido(x_data, y_data, x_omitido_index, x_omitido_value)
```
