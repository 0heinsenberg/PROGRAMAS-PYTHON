# Convert Data to Volts and Unwrap Phase

# dBm from Volts

$P_{dBm}=10\log_{10} \left( \frac{V^2}{R_l*P_{0}}\right) $

where $V$ is the voltage, $R_l$ is the load resistance, normally $50\Omega$, and $P_0$ is the power, $1mW$.

$\implies$ $P_{dBm}=20\log_{10}\left(V\right)-10\log_{10}\left(20\right)\simeq20\log_{10}\left(V\right)-13$ 

and

$V=10^{\frac{P_{dBm}-13}{20}}$

# Unwrap Phase

Unwrap by changing deltas between values to 2$\pi$ complement.

```python
# 10log(20) ≈ 13.01
c = 10*log10(20)
S21_Amp_l = pow(10,(S21_Amp-c)/20)#/1000

Calib_Amp_l = pow(10,(Calib_Amp-c)/20)#/1000

#Unwrap Phase

S21_Phase_l = unwrap(radians(S21_Phase))

Calib_Phase_l = unwrap(radians(Calib_Phase))
```

# Program to calculate S21 from Transmission Matrix

```python
def fitCPW(f, alpha, Cin,l, Ll, Cl,  Amp0=0, Phase0=0, Amp1=0, Phase1= 0,Rl= 50):
    """ Transmission Matrix function for CPW with loss. 
    Alpha = loss
    Cin = Cout = Coupling capacitor 
    l = Lenght resonator
    Ll = Inductance resonator withe Kinectic Inductance
    Cl = Capacitance resonator
    Amp0 & Phase0 = Background Amplitude and Phase 
    Rl = Impedance Source/Amplifier
    """
   
    Zin = 1/(1j*2*pi*f*Cin)
    
    Z0 = sqrt(Ll/Cl)
    
    floor = Amp0*exp(1j*Phase0)#complex(Amp0,Phase0)
    
    gamma = l*(alpha + 2j*f*pi*sqrt(Ll*Cl))
    sh = sinh(gamma)
    ch = cosh(gamma)
    
    A = ch + Zin*sh/Z0
    B = Zin**2*sh/Z0+ 2*Zin*ch+Z0*sh
    C = sh/Z0
    D = A
    
    S21 = 2/(A+B/Rl+Rl*C+D) + floor*f 
    
    return S21

def fitCPW2(f,S21_max,Q,f0,Amp0=0,Phase0=0):
    """ Complex fit using Lorentzian curve and error term.
    S21_max = Max value S21 linear 
    Q = loaded Quality Factor
    Amp0 = amplitude error
    Phase0 = Phase error
    """
    
    floor = Amp0*exp(1j*Phase0)#complex(Amp0,Phase0)
    
    S21 = S21_max/(1+1j*Q*(f/f0-f0/f)) + (floor)*f 

    return S21

```

# Real Imaginary Fit and rotation data

```python
# Method to calculate the center and radius of points in a circuference
from scipy      import  odr

# Import Data
X0 = S21_freq
Y0 = S21_Amp_l_Calib*(exp(1j*S21_Phase_l_Fit))

x = Y0.real
y = Y0.imag

x_m = mean(x)
y_m = mean(y)

def calc_R(xc, yc):
    """ calculate the distance of each 2D points from the center (xc, yc) """
    return sqrt((x-xc)**2 + (y-yc)**2)

x_m = mean(x)
y_m = mean(y)


# method_3 = "odr"

def f_3(beta, x):
    """ implicit definition of the circle """
    return (x[0]-beta[0])**2 + (x[1]-beta[1])**2 -beta[2]**2

# initial guess for parameters
R_m = calc_R(x_m, y_m).mean()
beta0 = [ x_m, y_m, R_m]

d = sqrt((x-x_m)**2+(y-y_m)**2)
error = R_m/sqrt(mean((d-R_m)**2))
# print(error)


# for implicit function :
#       data.x contains both coordinates of the points (data.x = [x, y])
#       data.y is the dimensionality of the response
lsc_data  = odr.Data(row_stack([x, y]),y=1)#,we=1/error,wd=error)
lsc_model = odr.Model(f_3, implicit=True)
lsc_odr   = odr.ODR(lsc_data, lsc_model, beta0)
lsc_out   = lsc_odr.run()

xc_3, yc_3, R_3 = lsc_out.beta
Ri_3 = calc_R(xc_3, yc_3)
residu_3 = sum((Ri_3 - R_3)**2)
lsc_out.beta
```

```python
x0 = lsc_out.beta[0]
y0 = lsc_out.beta[1]
r0 = lsc_out.beta[2]


a = linspace(0,2*pi,1000)
x = r0 + r0 *cos(a)
y =  r0 *sin(a)
print(x0,y0,r0)

plot(x,y,'g-')
plot(Y0.real,Y0.imag,'ro')
Angle = arctan2(y0,x0)
print(Angle)
Y1=(Y0-complex(x0,y0))*exp(-1j*Angle)+sqrt(x0**2+y0**2)
# Y1=(Y0)*exp(1j*Angle)

plot(Y1.real,Y1.imag,'kv')
# plot(Fit.real,Fit.imag,'go')
hlines(0,min(Y1.real),max(Y1.real))
vlines(0,min(Y1.imag),max(Y1.imag))
```

# Fit function S21 Matrix 

```python
def residual(pars, f, S21,Error):
    alpha, Cin,l, Ll, Cl,  Amp0, Phase0 = pars
#     diff_real = fitCPW(f, alpha, Cin,l, Ll, Cl).real - S21.real
#     diff_imag = fitCPW(f, alpha, Cin,l, Ll, Cl).imag - S21.imag
    diff = fitCPW(f, alpha, Cin,l, Ll, Cl,Amp0,Phase0) - S21
    return  (abs(diff*Error))#diff.real**2 + diff.imag**2#


```

```python
# initial values 
alpha, Cin,l, Ll, Cl,Amp0, Phase0, Rl = 1.034e-4, 1.9e-15,0.01103,4.343e-7,1.694e-10,0,0,50
par_init = alpha, Cin ,l, Ll, Cl,Amp0, Phase0
# S21_max, Q,f0 = 0.26790275119624102,50000,5.13661e9
# par_init_2 = S21_max,Q,f0,Amp0,Phase0
Error = Ri_3

# popt, pcov = curve_fit(fitCPW, x, y,par_init)
best, cov, info, message, ier = leastsq(residual, par_init, args=(X0,Y1,Error),\
                                        full_output=True,ftol=1e-20,xtol=1e-20,gtol=1e-9, maxfev=100000 )
# ftol=1e-30,xtol=1e-30,gtol=1e-30,
print(" Best-Fit Parameters: ",  best,cov)
```

# Quality factor

```python

alpha, Cin ,l, Ll, Cl,Amp0, Phase0= best
L = 2 * Ll*l/(pi*pi)
C = Cl*l/2
z0 = sqrt(Ll/Cl)
R = z0/(alpha*l)
F0 = 1/(2*pi*sqrt(L*C))
wn = 2*pi*F0
Qi = 2*pi*F0*R*C

a = (wn**2)*(Cin**2)*Rl
Rstar = (1+a*Rl)/a
Cstar = (Cin)/(1+a*Rl)
Wstar = 1/(sqrt(L*(C+2*Cin)))
fstar = Wstar/(2*pi) 
Qext = wn*Rstar*C/2
Qload = wn*(C+2*Cstar)/(1/R+2/Rstar)

print(Qi/1e3,Qext/1e3,Qload/1e3,fstar/1e9)
```
