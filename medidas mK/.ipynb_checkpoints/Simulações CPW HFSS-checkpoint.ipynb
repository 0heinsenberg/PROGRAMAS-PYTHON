## $F_{Lor}(f)=A_o\frac{\delta f}{(f-f_0)^2+(\delta f/2)^2}$

```python
f_max=np.max(S21_chip1_res3_abs_lor)

for j in range(0,len(S21_chip1_res3_abs_lor)):
    if S21_chip1_res3_abs_lor[j] == f_max:
        f_0=S21_freq_chip1_res3_lor[j]

print(f_max)
```

```python
#fitting Lorentziano

# Definição da função Lorentziana
def lorentzian(x, A, x0, deltaf):
    return A * ((deltaf)/((x-x0)**2+(deltaf/2)**2))

# Gerando dados de exemplo (para fins de demonstração)
np.random.seed(0)  # Para reprodutibilidade
x_data = S21_freq_chip1_res3_lor
y_data = S21_chip1_res3_abs_lor

# Ajustando a curva Lorentziana aos dados
initial_guess = [23, 6.3, 1]  # Palpite inicial para os parâmetros [A, x0, gamma]
params, covariance = curve_fit(lorentzian, x_data, y_data, p0=initial_guess)

# Extraindo os parâmetros ajustados
A_fit, x0_fit, deltaf_fit = params

# Gerando a curva ajustada
F_lor = lorentzian(x_data, A_fit, x0_fit, deltaf_fit)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20,12))

ax.plot(S21_freq_chip1_res3_lor, F_lor, linewidth=3, color='blue')
ax.plot(S21_freq_chip1_res3_lor, S21_chip1_res3_abs_lor, linewidth=3, color='black')
ax.set_title("Simulação Ansys CHIP1 Ressonador 3", fontsize=25)
ax.set_xlabel(r"Frequencia (GHz)", fontsize=25)
ax.set_ylabel(r"Transmissão (dB)", fontsize=25)
ax.tick_params(axis='both', which='major', labelsize=25)
#ax.set_xlim(6.2,6.5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Definição da função Lorentziana
def lorentzian(x, A, x0, gamma):
    return (A / np.pi) * (gamma / ((x - x0)**2 + gamma**2))

# Gerando dados de exemplo (para fins de demonstração)
np.random.seed(0)  # Para reprodutibilidade
x_data = S21_freq_chip1_res3_lor
y_data = S21_chip1_res3_abs_lor

# Ajustando a curva Lorentziana aos dados
initial_guess = [1, 0, 1]  # Palpite inicial para os parâmetros [A, x0, gamma]
params, covariance = curve_fit(lorentzian, x_data, y_data, p0=initial_guess)

# Extraindo os parâmetros ajustados
A_fit, x0_fit, gamma_fit = params

# Gerando a curva ajustada
y_fit = lorentzian(x_data, A_fit, x0_fit, gamma_fit)

# Plotando os dados e a curva ajustada
plt.plot(x_data, y_data, 'b-', label='Dados')
plt.plot(x_data, y_fit, 'r-', label='Ajuste Lorentziano')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import curve_fit
from scipy.stats import norm
import ipywidgets as widgets
from IPython.display import display

# Definindo a função Lorentziana com termo de amplitude
def lorentziana(x, amplitude, x0, gamma, offset):
    return amplitude * (gamma**2 / ((x - x0)**2 + gamma**2)) + offset

# Gerando dados de exemplo com ruído
def gerar_dados(amplitude, x0, gamma, offset, num_points=100, noise_level=0.05):
    x_data = np.linspace(-10, 10, num_points)
    y_data = lorentziana(x_data, amplitude, x0, gamma, offset)
    noise = noise_level * np.random.normal(size=x_data.size)
    y_data_noisy = y_data + noise
    return x_data, y_data_noisy

# Ajustando a Lorentziana aos dados
def ajustar_lorentziana(x_data, y_data):
    initial_guess = [1.0, 0.0, 1.0, 0.0]
    popt, pcov = curve_fit(lorentziana, x_data, y_data, p0=initial_guess)
    perr = np.sqrt(np.diag(pcov))  # Erros padrão dos parâmetros
    return popt, perr

# Análise dos resíduos
def analisar_residuos(x_data, y_data, popt):
    y_fit = lorentziana(x_data, *popt)
    residuos = y_data - y_fit
    return residuos

# Plotando os dados e o ajuste com resíduos
def plotar_resultados(x_data, y_data, popt, perr):
    y_fit = lorentziana(x_data, *popt)
    residuos = analisar_residuos(x_data, y_data, popt)
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))
    
    # Plot dos dados e ajuste
    ax1.scatter(x_data, y_data, label='Dados com ruído', color='blue')
    ax1.plot(x_data, y_fit, label='Ajuste Lorentziana', color='red')
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.legend()
    
    # Plot dos resíduos
    ax2.scatter(x_data, residuos, label='Resíduos', color='green')
    ax2.axhline(0, color='black', linestyle='--')
    ax2.set_xlabel('x')
    ax2.set_ylabel('Resíduo')
    ax2.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Intervalos de confiança dos parâmetros ajustados
    print("Parâmetros ajustados e intervalos de confiança (95%):")
    for i, param in enumerate(['Amplitude', 'x0', 'Gamma', 'Offset']):
        print(f"{param}: {popt[i]:.3f} ± {1.96 * perr[i]:.3f}")

# Interface gráfica interativa
def interface_interativa(amplitude, x0, gamma, offset, num_points, noise_level):
    x_data, y_data_noisy = gerar_dados(amplitude, x0, gamma, offset, num_points, noise_level)
    popt, perr = ajustar_lorentziana(x_data, y_data_noisy)
    plotar_resultados(x_data, y_data_noisy, popt, perr)

# Widgets para ajuste interativo
amplitude_slider = widgets.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='Amplitude:')
x0_slider = widgets.FloatSlider(value=0.0, min=-5.0, max=5.0, step=0.1, description='x0:')
gamma_slider = widgets.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='Gamma:')
offset_slider = widgets.FloatSlider(value=0.1, min=-1.0, max=1.0, step=0.1, description='Offset:')
num_points_slider = widgets.IntSlider(value=100, min=50, max=500, step=10, description='Num Points:')
noise_level_slider = widgets.FloatSlider(value=0.05, min=0.0, max=0.2, step=0.01, description='Noise Level:')

widgets.interactive(interface_interativa, 
                    amplitude=amplitude_slider,
                    x0=x0_slider,
                    gamma=gamma_slider,
                    offset=offset_slider,
                    num_points=num_points_slider,
                    noise_level=noise_level_slider)
```
