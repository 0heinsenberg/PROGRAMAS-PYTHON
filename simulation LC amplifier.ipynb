```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.interpolate import interp2d
import numpy as np
import matplotlib.cm as cm
from matplotlib.colors import LogNorm
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show



#CONFIG CIRCUITO
L= 50*(10**-3) #uH
C= 4.2*(10**-9) #nF
t=np.linspace(0,0.0005,100)
I0=1
w0= 1/(np.sqrt(L*C))
f0= w0/(2*np.pi)
print(f0, 'Hz')

I1=I0*np.sin(w0*t)

#GRÁFICO SEM AMPLIFICAÇÃO
plt.plot(t, I1)
plt.title('SEM AMPLIFICAÇÃO')
plt.ylabel('Corrente [A]')
plt.xlabel('tempo [s]')
plt.show()

#AMPLIFICADO
epsilon=0.01
fp=2*f0 #frequencia da pump
Lmod=L+(epsilon*np.sin(2*np.pi*fp*t))
w1=1/(np.sqrt(Lmod*C))

I2=I0*np.cos(w1*t)

#GRÁFICO COM AMPLIFICAÇÃO
plt.plot(t, I2)
plt.title('COM AMPLIFICAÇÃO')
plt.ylabel('Corrente [A]')
plt.xlabel('tempo [s]')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.interpolate import interp2d
import numpy as np
import matplotlib.cm as cm
from matplotlib.colors import LogNorm
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show

t=np.linspace(0,10,10000)
t2=np.linspace(0,10,10000)

L= 0.1 #uH
C= 0.1 #nF
w0=1/(np.sqrt(L*C))
f0=w0/(2*np.pi)
print(f0)
I0=0.005 #A
I1=I0*(np.cos(w0*t))

#GRÁFICO SEM AMP
plt.plot(t, I1)
plt.title('SEM AMPLIFICAÇÃO')
plt.ylabel('Corrente [A]')
plt.xlabel('tempo [s]')
plt.show()

delta=0.01
Lamp=L+(delta*(np.cos(2*w0*t2)))
w1=1/(np.sqrt(Lamp*C))
I2=I0*(np.cos(w1*t2))

#GRÁFICO COM AMP
plt.plot(t2, I2)
plt.title('COM AMPLIFICAÇÃO')
plt.ylabel('Corrente [A]')
plt.xlabel('tempo [s]')
plt.show()
```
