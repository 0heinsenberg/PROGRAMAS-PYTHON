## Amplificação Clássica 

```python
import matplotlib.pyplot as plt
from scipy.interpolate import interp2d
import numpy as np
import matplotlib.cm as cm
from matplotlib.colors import LogNorm
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
from qutip import *
```

```python
#parâmetros
g=10 #gravidade
l=1000 #comprimento do pendulo
ws=np.sqrt(g/l) #oscilação do pendulo
theta0=np.sqrt(2)/2 #angulo inicial
w0=100 #frequencia inicial
```

```python
t=np.linspace(0,2,10000)

```

## Solução com ws constante 

```python
theta1=theta0 * (np.cos(ws*t)) + (w0/ws) * (np.sin(ws*t))

###gráfico
plt.plot(t, np.abs(theta1))
plt.ylabel(r'$\theta_1$ [Graus]')
plt.xlabel(r'$tempo$ [s]')
plt.show()
```

## Solução com modulação em wCM=2ws 

```python
delta=1
wst=w0+(delta*np.sin(2*ws*t))
theta2=(theta0*(np.exp(0.5*delta*t))*np.cos(wst*t))+(w0*(np.exp((-0.5)*delta*t))*np.sin(wst*t))
```

```python
###gráfico
y=theta0*(np.exp(0.5*delta*t))
plt.plot(t, np.abs(theta2))
plt.plot(t,y)
plt.ylabel(r'$\theta_2$ [Graus]')
plt.xlabel(r'$tempo$ [s]')
plt.show()
```

```python
ganho=theta2/theta1


###gráfico
plt.plot(t, ganho)
plt.ylabel(r'$Ganho$ [Graus]')
plt.xlabel(r'$tempo$ [s]')
plt.show()
```

# Relação VxI JJ

```python
Ic=1
Rj=1
xmais=np.linspace(1,2,1000)
x0=np.linspace(-1,1,100)
Vmed0=Ic*x0*0
Vmed_pos=Ic*Rj*(np.sqrt((xmais)**2-1))
Vmed_neg=-Vmed_pos
xdash=np.linspace(-2,2,1000)
Vohmic=Rj*xdash

#gráfico
plt.plot(xmais,Vmed_pos, color="blue")
plt.plot(-xmais,Vmed_neg, color="blue")
plt.plot(x0,Vmed0, color="red")
plt.plot(xdash,Vohmic, color="black", linestyle="--")
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel(r'$V_{med}$ [V]', fontsize="20")
plt.xlabel(r'$Corrente$ [A]', fontsize="20")
plt.show()
```
